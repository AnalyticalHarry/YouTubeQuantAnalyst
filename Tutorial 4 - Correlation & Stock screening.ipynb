### US MARKET

The S&P 500 can have a significant impact on the global economy due to the sheer size and influence of the US economy, which is the largest in the world. The performance of the S&P 500 can affect not only the US economy but also other countries and regions that are closely tied to the US economy through trade, investment, and other forms of economic activity.

When the S&P 500 performs well, it can boost investor confidence and lead to increased investment in the US economy, which can, in turn, stimulate economic growth and job creation. A strong US economy can also increase demand for goods and services from other countries, thereby supporting economic growth around the world.

Conversely, when the S&P 500 performs poorly, it can have negative ripple effects on the global economy. A decline in the US stock market can reduce investor confidence and lead to a reduction in investment and economic activity, both in the US and in other countries that are closely tied to the US economy.

Moreover, the S&P 500 can serve as a bellwether for broader trends in the global economy. Changes in the index can reflect shifts in consumer behavior, changes in the regulatory environment, and other factors that can have far-reaching impacts on the global economy.

The symbols you mentioned are related to the S&P 500 index, which is a stock market index that measures the performance of 500 large companies listed on stock exchanges in the United States. Here's the breakdown of each symbol:

^GSPC: This is the Yahoo Finance symbol for the S&P 500 index. The caret symbol (^) is often used to represent an index in financial platforms.

$SPX: This is the symbol used by some financial platforms, including Bloomberg and CNBC, to represent the S&P 500 index. The dollar sign ($) is commonly used to denote an index.

SPX: This is the ticker symbol used on options and futures contracts for the S&P 500 index. It represents a specific contract related to the index.

.SPX: In some financial platforms, including Reuters and Google Finance, a dot (.) is used as a prefix to represent the S&P 500 index. This is similar to how some platforms use a dot for international stock indices (e.g., .DJI for the Dow Jones Industrial Average).

.INX: This is another symbol used by some financial platforms, such as Yahoo Finance and Google Finance, to represent the S&P 500 index. It is an abbreviation of "Index" and is used to differentiate the index from individual stocks.

## CORRELATION ANALYSIS

```python
import ta
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
corr_matrix = us_price_data.corr()
```

```python
corr_matrix
```

```python
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)
```

```python
corr_pairs
```

```python
corr_pairs = corr_pairs[corr_pairs !=1 ].drop_duplicates()[:10]
corr_pairs = pd.DataFrame(corr_pairs).reset_index()
```

```python
corr_pairs
```

```python
sns.set_style("white")
plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix.loc[corr_pairs.iloc[:, 0], corr_pairs.iloc[:,1]],cmap="Greens", annot=True, fmt=".2f")
plt.show()
```

```python
us_price_data.columns
```

```python
rsi_data = pd.DataFrame()
for col in us_price_data.columns:
    rsi = ta.momentum.RSIIndicator(us_price_data[col]).rsi()
    rsi_data[col] = rsi
```

```python
rsi_data.dropna(inplace=True)
```

```python
rsi_data
```

```python
low_rsi = rsi_data.iloc[-1] < 30
```

```python
low_rsi
```

```python
safe_companies = low_rsi.sort_values().index[:10]
```

```python
safe_companies
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(15,6))
us_price_data[safe_companies].plot(kind="area", alpha=0.5, stacked=False, figsize=(10,6))
plt.ylabel("Price ($)")
plt.title("Top ten less volaitle Investment")
plt.show()
plt.close()
```

## TOP TEN HIGHEST SHARE PRICE

```python
high_price_companies = us_price_data.max().sort_values(ascending=False).index[:10]
```

```python
high_price_companies
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(15,6))
us_price_data[high_price_companies].plot(kind="area", alpha=0.5, stacked=False, figsize=(10,6))
plt.ylabel("Price ($)")
plt.title("Top ten highest price stock")
plt.grid(linestyle="--", color="white", alpha=0.7)
plt.show()
plt.close()
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(15,6))
us_price_data[high_price_companies].mean().plot(kind="bar", alpha=0.5, stacked=False, figsize=(10,6), color="red")
plt.ylabel("Price ($)")
plt.title("Top ten highest price stock")
plt.grid(linestyle="--", color="white", alpha=0.7)
plt.show()
plt.close()
```
