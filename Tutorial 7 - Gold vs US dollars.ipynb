```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class stock:
    def __init__(self, ticker, period):
        self.ticker = ticker
        self.period = period

    def chart(self):
          return yf.download(self.ticker, period=self.period)
```

The U.S. dollar index (USDX) is a measure of the value of the U.S. dollar relative to a basket of foreign currencies. The USDX was established by the U.S. Federal Reserve in 1973 after the dissolution of the Bretton Woods Agreement. It is now maintained by ICE Data Indices, a subsidiary of the Intercontinental Exchange (ICE).

The U.S. Dollar Index (USDX) measures the value of the U.S. dollar against a basket of six major foreign currencies. These six currencies are:

1. Euro (EUR)
2. Japanese Yen (JPY)
3. British Pound Sterling (GBP)
4. Canadian Dollar (CAD)
5. Swedish Krona (SEK)
6. Swiss Franc (CHF)

## U.S DOLLAR INDEX 

```python
usd = stock("DX-Y.NYB", "MAX").chart()
```

```python
plt.figure(figsize=(15,6))
plt.plot(usd.index, usd['Adj Close'], alpha=0.2)
usd['Adj Close'].plot(color='grey')
plt.ylabel('USD ($)')
plt.grid(axis='y',linestyle='--')
plt.show()
```

The U.S. Dollar Index has risen and fallen sharply throughout its history. It reached an all-time high in 1984 at nearly 165. Its all-time low was nearly 70 in 2007. Over the last several years, the U.S. dollar index has been relatively rangebound between 90 and 110.

The index is affected by macroeconomic factors, including inflation/deflation in the dollar and foreign currencies included in the comparable basket, as well as recessions and economic growth in those countries.

```python
intial_usd = usd['Adj Close'][0]
final_usd = usd['Adj Close'][-1]
usd_growth = (final_usd - intial_usd) / intial_usd * 100
print(f'USD Growth: {round(usd_growth,2)} %')
```

```python
intial_inr = inr['Adj Close'][0]
final_inr = inr['Adj Close'][-1]
inr_growth = (final_inr - intial_inr) / intial_inr * 100
print(f'INR Growth: {round(inr_growth,2)} %')
```

```python
plt.figure(figsize=(15,6))
plt.plot(usd.index, (usd['Adj Close']/ usd['Adj Close'].iloc[0] - 1)*100, label="USD")
plt.plot(inr.index, (inr['Adj Close']/ inr['Adj Close'].iloc[0] - 1)*100, label="INR")
plt.ylabel('Growth percentage (%)')
plt.grid(axis='y',linestyle='--')
plt.title("USD & INR")
plt.legend()
plt.show()
```

```python
gcf = stock("GC=F", "5Y").chart()
```

```python
plt.figure(figsize=(15,6))
gcf.Close.plot()
plt.ylabel('USD')
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
intial_gcf = gcf['Adj Close'][0]
final_gcf = gcf['Adj Close'][-1]
gcf_growth = (final_gcf - intial_gcf) / intial_gcf * 100
print(f'GOLD Growth: {round(gcf_growth,2)} %')
```

```python
plt.figure(figsize=(15,6))
plt.plot(gcf.index, (gcf['Adj Close']/gcf['Adj Close'].iloc[0]-1)*100, linewidth=1)
plt.ylabel('Growth percentage (%)')
plt.grid(axis='y',linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.plot(usd.index, (usd['Adj Close']/ usd['Adj Close'].iloc[0] - 1)*100, label='USD')
plt.plot(gcf.index, (gcf['Adj Close']/gcf['Adj Close'].iloc[0]-1)*100, label="GOLD")
plt.ylabel('Growth percentage (%)')
plt.grid(axis='y',linestyle='--')
plt.title('GOLD & USD')
plt.legend()
plt.show()
```

The price of gold and the value of the U.S. dollar typically have an inverse relationship. When the U.S. dollar strengthens relative to other currencies, gold prices tend to decrease, and when the dollar weakens, gold prices often rise. This relationship can be explained by factors such as currency exchange rates, the role of gold as a safe-haven asset, interest rates, market sentiment, and the self-fulfilling nature of this correlation. However, it's important to note that the relationship is not absolute, and other factors can also influence the price of gold. Investors and traders often monitor this relationship, but they consider various factors when making investment decisions involving gold or the U.S. dollar.
