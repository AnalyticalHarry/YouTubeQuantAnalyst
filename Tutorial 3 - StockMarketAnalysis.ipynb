```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
import yfinance as yf
```

## MSFT

```python
stock1 = yf.Ticker("MSFT")
```

```python
stock1
```

```python
stock1_history = stock1.history(period="10y")
```

```python
msft = stock1_history
msft.shape
```

```python
msft
```

```python
msft.Dividends.unique()
```

```python
msft.Dividends.value_counts()
```

```python
msft.Dividends.value_counts().plot.bar()
```

```python
msft.columns
```

```python
msft.columns = msft.columns.str.lower()
```

```python
msft.columns
```

```python
msft.isnull().sum()
```

```python
plt.figure(figsize=(15,6))
msft.open.plot(color="red")
msft.close.plot(color="royalblue")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Microsoft")
plt.grid(linestyle="--", color="grey")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
msft.high.plot(color="orange")
msft.low.plot(color="green")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Microsoft")
plt.grid(linestyle="--", color="grey")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
msft.open.tail(365).plot(color="red")
msft.close.tail(365).plot(color="royalblue")
msft.high.tail(365).plot(color="orange")
msft.low.tail(365).plot(color="green")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Microsoft")
plt.grid(linestyle="--", color="grey")
plt.legend()
plt.show()
```

## AAPL

```python
stock2 = yf.Ticker("AAPL")
stock2_history = stock2.history(period="10y")
aapl = stock2_history
aapl.shape
```

```python
aapl.columns = aapl.columns.str.lower()
```

```python
aapl
```

```python
aapl.columns
```

```python
plt.figure(figsize=(15,6))
aapl.close.plot(color="royalblue")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Apple")
plt.grid(linestyle="--", color="grey")
plt.legend()
plt.show()
```

## Comparsion between Apple & Microsoft

```python
plt.figure(figsize=(15,6))
msft.close.plot(color="red", label="Microsoft")
aapl.close.plot(color="blue", label="Apple")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Apple vs Microsoft")

plt.axhline(msft.close.mean(), linestyle="--", color="red")
plt.axhline(aapl.close.mean(), linestyle="--", color="blue")

plt.grid(linestyle="--", color="grey")
plt.legend()
plt.show()
```
