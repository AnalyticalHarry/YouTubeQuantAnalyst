### ECONOMIC PERFORMANCE & INVESTMENT OPPORTUNITY

```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def calculate_profit(investment, intial_price, future_price):
    initial_value = investment/intial_price
    future_value = initial_value * future_price
    profit = future_value - investment
    return profit
```

```python
def plot_profit_over_time(dates, profits):
    plt.style.use('dark_background')
    plt.figure(figsize=(15, 6))
    plt.plot(dates, profits, linewidth=2, color="gold")
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Profit', fontsize=18)
    plt.title('Profit Over Time')
    plt.grid(linestyle="--", color="white", alpha=0.7)
    plt.xticks(rotation=45)
    plt.show()
    plt.close()
```

```python
total_investment = 10_000
```

```python
intial_share_price = prices.iloc[0]
```

```python
future_Share_price = prices
```

```python
profits = []
for price in future_Share_price:
    profit = calculate_profit(total_investment, intial_share_price, price)
    profits.append(profit)
```

```python
len(profits)
```

```python
final_profit = profits[-1]
```

```python
final_amount = total_investment + final_profit
percentage_change = (final_profit/total_investment)*100
```

```python
plot_profit_over_time(dates, profits)
```
